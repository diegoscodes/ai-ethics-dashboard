```python
# 01_data_exploration.ipynb
# --------------------------
# initial data exploration for AI Ethics Dashboard project

import pandas as pd

# Load dataset
df=pd.read_csv("data/raw/adult.csv")

# Quick overview
print("shape:", df.shape)
display(df.head())

# info and missing values
df.info()
print("\nMissing values per column:")
print(df.isna().sum())
```

```python
# Count target values
print(df['income'].value_counts())

# Gender distribution
print(df['sex'].value_counts())

# Race disttribution
print(df['race'].value_counts())
```

```python
import matplotlib.pyplot as plt

df['sex'].value_counts().plot(kind='bar', color=['royalblue', 'lightcoral'])
plt.title("Gender distribution")
plt.show()
```
